```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ------------------------ Carga e información básica del dataframe a trabajar, filtrado de información ----------------------------------

df =  pd.read_csv("afluencia.csv", encoding= "ISO-8859-1") # Carga del csv dentro de un dataframe usando pandas
```

```python
# ------------------------- Graficado de información: Histogramas ----------------------------------------

fig, axs= plt.subplots(1,2, figsize=(20,6))  # Inicialización de la figura con subdivisiones

sns.histplot(data = df, x = 'anio', y = 'afluencia' ,bins=10, kde="True",ax = axs[0])   # Histograma
sns.histplot(data = df, x = 'linea', y = 'afluencia' ,bins=10, kde="True",ax = axs[1])
```

```python
# ------------------------------- Graficado de información: Mapas de calor ----------------------------------------

df2 = df[['anio','afluencia']] # Creación de dos dataframes que contengan únicamente las columnas a relacionar para los heatmaps
df3 = df[['linea','afluencia']]

fig, axs = plt.subplots(1,2,figsize=(16,4))  # Inicialización de la figura con subdivisiones

corr = df2.corr()  # Inicialización de la correlación para poder generar los mapas de calor
corr_ = df3.corr()

sns.heatmap(data = corr, vmin=-1, vmax=1, annot=True, cmap="winter", fmt = '.2f', ax = axs[0]) # Declaración de las funciones para graficar los mapas
plt.xlabel('anio')
plt.ylabel('afluencia')

sns.heatmap(data = corr_, vmin=-1, vmax=1, annot=True, cmap="cool", fmt = '.2f', ax = axs[1]) # Declaración de las funciones para graficar los mapas
plt.xlabel('linea')
plt.ylabel('afluencia')
```

```python
# ----------------------------- Graficado de información:Boxplot -----------------------------------

fig, axs = plt.subplots(1,2,figsize=(20,6))

sns.boxplot(data = df,x="anio", y="afluencia", ax = axs[0])
sns.boxplot(data = df,x="linea", y="afluencia", ax = axs[1])

```
